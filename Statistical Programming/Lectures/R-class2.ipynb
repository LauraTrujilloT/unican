- Make 4 plots for t student(2 by 2) w data between -3 and 3
    - Density functions for 1,5,20,30 degrees of freedom (df) + curve normal distribution (mean =0,sd=1)
    - Random numbers (1000) + histogram + t-student curve
    - Empirical cumulative ditribution function (ecdf) w random simulated data
    - Theoretical cumulative distribution function(CDF) 

```r
x <- seq(-3,3,length=100)
y1 <- dt(x, df=1)
y2 <- dt(x, df=5)
y3 <- dt(x, df=20)
y4 <- dt(x, df=30)
```

```r
plot.new()
plot(x, y4,col="red", main="T Student Density Functions", xlab=)
lines(x,y3, col="green")
lines(x,y2, col="blue")
lines(x,y1, col="magenta")
lines(x, dnorm(x,mean=0,sd=1), col="black", lw=2)
legend("topright", pch=c(1, NA, NA, NA,NA),legend=c("df=30","df=20", "df=5","df=1","Normal"), 
       lty=c(NA,1,1,1,1), col=c("red","green","blue","magenta","black"))
```
