### Example: Roots, Max & Min, Fitting
$$f(x) = 3 + 2x + x^{2} = 0 $$

```r
options(repr.plot.width = 4, repr.plot.height=3)
# Polynomial :
pol1 <- function(x) 3 + 2*x + x^2

#Plotting
curve(pol1, xlim=c(-100,100), cex.axis=0.8, cex.lab=0.8) 
#cex.axis reduce axis since we're changing size of plot. Also labels

#Roots
x <- seq(-100, 100, by=0.1)
cat("Is it negative?", any(pol1(x)<0), "\n")
zeros <- polyroot(c(3,2,1))   # look for roots
print(zeros)                  # no real roots; only complex roots
```

### Exercise: roots, plot, points for roots
$$f(x) = (x-3)(x - \frac{1}{2})(x+\frac{1}{2})(x+4) = 3 - \frac{1}{4}x - \frac{49}{4}x^2 + x^3 + x^4 $$

```r
options(repr.plot.width = 4, repr.plot.height=4)
pol2 <- function(x) 3 - (1/4) * x - (49/4) * x^2 + x^3 + x^4

#Plot
curve(pol2, xlim=c(-5,5), ylim = c(-50,50), col="red",  cex.axis=0.8, cex.lab=0.8, lwd=2,
     main=expression((x-3)(x - frac(1,2))(x + frac(1,2))(x+4)))
grid()
abline(h=0, col="green",lty=2)

zeros2 <- polyroot(c(3, -1/4, -49/4, 1, 1)) 
print(zeros2)

points(Re(zeros2), rep(0, 4), pch=16, cex=1.5)
```
