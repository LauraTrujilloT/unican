```r
load(file="muestra6.rda")
nc=5 #This only can vary from 1 til 5
trai=training
Values <- trai[,1:nc-1]
Targets <- trai[,nc]
```

```r
toyFunction <- function(LP1, LP2, max=1000, ratio=0.15){
    # LP1 = Learning parameter 1 (Learning rate)
    # LP2 = Learning parameter 2 (momentum)
    # max = Max number of iterations
    # ratios = ratio between values and targets of training 
    
    trai <- splitForTrainingAndTest(Values, Targets, ratio=0.15) #original ratio = 0.15
    str(trai)
    model <- mlp(trai$inputsTrain,trai$targetsTrain,size=c(20,20), #Original size 20,20
             inputsTest = trai$inputsTest, targetsTest = trai$targetsTest,
             outputActFunc = "Act_Logistic",
#             learnFunc = "Std_Backpropagation",learnFuncParams = c(0.1),
             learnFunc = "BackpropMomentum",learnFuncParams = c(LP1, LP2),  #original 0.1, 0.1
#             learnFunc = "SCG", learnFuncParams = c(0.1),
              maxit = max) #maxit original = 1000
#summary(model)
#plotnet(model)
    readline(prompt="Press [enter] to continue")
    plotIterativeError(model)
    readline(prompt="Press [enter] to continue")
    sum(model$fitted.values[Targets==0]>0.5)
    plotROC(model$fitted.values,Targets)
    readline(prompt="Press [enter] to continue")
    attributes(model)
    hist(model$fitted.values[Targets==0],col=rgb(1,0,0,0.3))
    hist(model$fitted.values[Targets==1],col=rgb(0,0,1,0.3),add=T)
    readline(prompt="Press [enter] to continue")
    readline(prompt="Press [enter] to continue")
    par(mfrow = c(2, 2),     # 2x2 layout
        oma = c(0, 0, 0, 0), # two rows of text at the outer left and bottom margin
        mar = c(1, 1, 0, 0), # space for one row of text at ticks and to separate plots
        mgp = c(2, 1, 0),    # axis label at 2 rows distance, tick labels at 1 row
        xpd = NA)            # allow content to protrude into outer margin (and beyond)
    plot(trai$inputsTest[,1],predict(model,trai$inputsTest),xlab="",ylab="")
    plot(trai$inputsTest[,2],predict(model,trai$inputsTest),xlab="",ylab="")
    plot(trai$inputsTest[,3],predict(model,trai$inputsTest),xlab="",ylab="")
    plot(trai$inputsTest[,4],predict(model,trai$inputsTest),xlab="",ylab="")
    readline(prompt="Press [enter] to continue")
    par(mfrow=c(1,1))
}
```

```r
toyFunction(0.3, 0.7, 3000)
```

```r
toyFunction(0.1, 0.4, 3000)
```

```r
toyFunction(0.1, 0.1, 10000)
```

## Observations
- Changing for greater values than $0.1$ of _learning rate_ leads to divergence (overtraining)
- Also overtraining when changing the momentum and making the learning rate $0.1$. Seems like it "stops learning" quicker when changing the learning rate than when changing the momentum.
- When the number of iterations goes greater than $2000$ and both learning parameters are at 0.1, the model starts to be overtrained as well.


For the `plotnet` it was obtained the following figure for `LP1=0.1, LP2 = 0.1, max = 1000, ratio = 0.15`
![Rplot.png](Rplot.png)

The best scenario could be one with number of iterations less than $1000$ or $2000" which is found to be when the model stops learning
