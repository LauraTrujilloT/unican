## Statistical Functions

See [description](http://venus.ifca.unican.es/Rintro/statisticalTreatment.html) at Rintro tutorial.

Make four plots for t-Student (2 by 2) with data between -3 and 3 and representing:

1. Density functions for 1, 5, 20, 30 degrees of freedom (df) + curve for normal distribution (mean=0, sd=1)
2. Random numbers (1000) + histogram + t-Student density curve
3. Empirical Cumulative Distribution Function (ECDF) with the random simulated data
4. Theoretical Cumulative Distribution Function (CDF)

````r
par(mfrow=c(2,2))
# a) density function
x<-seq(-3.,3,length=100)

plot(x,dt(x,df=30),type="l",col="cyan",ylab = "Density Function (x)", main="t-Student density function")
lines(x,dt(x,df=20),col="blue")
lines(x,dt(x,df=5),col="green")
lines(x,dt(x,df=1),col="magenta")
points(x,dnorm(x,mean=0,sd=1), cex=0.3, pch=4)
legend("topleft",c("Distr. t with df=1","Distr. t with df=5", "Distr. t with df=20",
                   "Distr. t with df=30", "Distr. Normal(0,1)"),
       col=c("magenta","green","blue","cyan","black"),cex=0.6,
       lty=c(1,1,1,1,NA),pch=c(NA,NA,NA,NA,21),bty="n",y.intersp = 0.5)

# b) random numbers
simdata<-rt(1000,df=30)
#hist(simdata, xlab="random t Student", main="Random data (rdmData)")
#rug(simdata)
hist(simdata, xlab="random t Student", main="Random data (rdmData)", freq=FALSE)
curve(dt(x,df=30),add=T,col="cyan")


# c) ECDF
plot(ecdf(simdata),verticals = TRUE, col.points = "blue", col.hor = "red", 
     col.vert = "bisque",xlab="random t Student", ylab="ECDF(x)", 
     main="ECDF(rmdData)", xlim=c(min(simdata),max(simdata)))

# d) Th CDF
plot(simdata,pt(simdata,df=30),ylab="CDF(x)",xlab="random t Student",cex=0.3, 
     main="CDF: pt(rmdData, df=30)",xlim=c(min(simdata),max(simdata)))
````

## Deliverable Ex. 2

Data: dat1.dat, dat2.dat, dat3.dat

Use several test seen in class to compare those data distributions

1. Do they have the same mean?
2. Do they have the same variance?
3. Are they compatible with being the same distribution?

Ho = both distributions have same mean/variance (the *interesting thing* is to be able to reject Ho...)

HA = (Alternative): both distributions have different mean/variance (mu1>mu2 o mu2>mu1)

P(Ho)= probability of being wrong if we reject Ho ~ probability of both distributions having same mean/variance...:

         * if they are equal P(Ho) ~ 1
         * if they are different P(Ho)~ 0 (I can reject Ho)

P(Ho) = *.test$p.value

P(Ho) = 2 * pt,f,..(t|f|?, nu,...)

**TESTS:**

    * t (Student) test: parametric test for the mean
        Statistic: t 
        Dof: nu 
        Two cases (sigmas are equal/different - aprox. Welch)
        Use t.test &&  2*pt (t,nu,..., inequality test)
        
        t.test(x, y = NULL, alternative = c("two.sided", "less", "greater"), mu = 0, paired=FALSE, 
                var.equal = FALSE, conf.level = 0.95)
        Then the null hypothesis of the two-tailed test is to be rejected if t ≤−tα∕2 or t ≥ tα∕2
        
    * F (isher) test: parametric test for the variance
        Use var.test  &&  2*pf (F, nu1, nu2,...,inequality test)
        Statistic: F
        Dof: nu1=N1-1; nu2=N2-1
        
    * Non-parametric test of the mean: t by ranks (rank summatory)     
        Use t.test(on ranks) && wilcox.test (test U from Wilcoxon-Mann-Whitney over original data)
        
    * Z test : Tests (parametric) for the variance for large N
        Use Z (from expression) && z.test
        Calculate Z=(SA -SB)/(SD/sqrt(2))
        P(HA) = erf(Z)
        P(Ho) = 1 - P(HA) = erfc(Z)
        
        We cannot reject H0 (equal variance) if Z~0
        
    * Test F over ranks ranks: Variance tests for large N
        Use P(H0)= var.test$p.value (over ranks)
                
    * K-S test: compare the full distribution
        Use ks.test$p.value
