## Metropolis algorithm

A simplified version of the procedure, to sample $\rho(r)$, r representing a vector in n-dimensions.

1. We start by any allowed point r, so-called _walker_
2. We shake the point, get a new $r^\prime = r + \delta$ where $\delta$ is a vector such that each of its components is a uniform random number in [-a,a] (_random walk_
3. We evaluate $q= \rho(r^\prime) / \rho(r)$, the ratio of probabilities of new and original position.
4. Accept/reject $r^\prime$ with probability q
    - Draw a $t \sim U(0,1) $
    - If $t<q$ we make $r_{i+1} = r^\prime$, otherwise we keep $r_{i+1}=r$
    - Obviously if $q>1$ we always keep the new point
5. If we repeat this procedure r converges to a random number of $\rho(r)$
    - Note we itereate several times to get ONE number

```r
#First CREATE RHO function
N = 1e4
rho <- function(x){
    return(0.5 * dnorm(x, -2, 1) + 0.5 * dnorm(x, 2, 1))
}
```

```r
r0 <- runif(N, -2, 2)
delta <- runif(N, -3, 3)
rprime <- r0 + delta
q <- rho(rprime) / rho(r0)
```

```r
t <- runif(N, 0, 1)
r <- ifelse(t < q, rprime, r0)
```

```r
hist(r)
```

## 1. MC Integration
- Understangin the integral as a sum
- Can be improved and extended to infinite  with importance sampling techniques
- Trivially extended to N dimensions and to complex hypervolumes
- Convergence doesnot depend on the dimensions or complexity of the problem

```r
N <- 1e4
m <- 1e3
b <- c(1:m)

for(i in 1:m){
    x <- runif(N, 0, 1)
    z <- x ^4
    b[i] <- mean(z)
}
```

```r
hist(b)
```

## 3. MC-based inference
- Related to the so-called generative models
- For example, we have a pdf that depends on a parameters $\theta$ we are interested in POI (Parameter of Interest):
    - We cannot model it analytically, but we know how to sample
    - We can fit, build a likelihood
    - Or compare through a gof (KS, chi square ..)
- We can solve it by
    - Generating a random sample according to a value of $\theta$ and calculating the statistics
    - Repeat for other values of $\theta$ and get the optimal one

- Start for a value of my $\theta$
- Generate random sample ${x_i}$
- Compare with data

**Example**: Given a counting experiment whose pdf is driven by a mixture model with one parameter (Poisson, whose mean follows a law $1 + \exp{(-\lambda x)}$ 

$\lambda$ such that $<n> = n_D$

```r
lambda <- function(N, nd){
    return(1 + rexp(N, nd))
}
```

```r
N <- 1e4
nd  <- 3
l1 <- lambda(N, nd)
x <- rpois(N, l1)
mu <- lambda(N, 3)
x <- rpois(N, mu)
hist(mu)
```

```r
hist(x)
```

```r
mean(x)
```

```r
mean(rpois(N, lambda(N, nd)))
```

## 4. p-values and pulls
- Calculate a p-value is trivial once one has an observation and a random sample of that pdf , just count how mant cases are above the observation 
- It works regardless of the complexity of the pdf, once we have the random sample.
- A usual application, when properly defina a "pull":
    - Pull is the jargon name for the z-score, equivalent number of "gaussian sigmas"
    - Widely used for visual or quick checks
    - Often defined as (data-prediction)/sigma

## 5. Application to discovery or exclusion
- A keysituation in many astrophysics or particle physics experiments.
- We perform an analysis (can be a single quantity or a spectrum)
- We have a prediction fiven by a model of known physics and another depending on new physics that we want to test, often depending on a parameter (for example the signal strength, how much additional physics we put in)
- We look for:
    - Discovery, proof that our data is not statistically compatible with the known physics. Basically find that the p-value for the known physics (KP) hypothesis is very **small**
    - Reject the model. Basically find that the p-value for the new physics (NP) hypothesis is **small**
    - Bound the allowed paramter of the new physics. Basically repeat b as a function of the parameters.
- In practice, this must be done by MC unless we have sufficiently large number of events and can use asymptotics (Wilks theorem)
