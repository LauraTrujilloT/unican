### Shell Basics 

- `ls`: Some flags $\rightarrow$ `-l`(w date), `-S`(Order by size), `-h`(human readable), `-r`(smallest to highest), 
    `-t`(order by date). You can also put together like `-lhtr`. 
 
 **example** : `ls -w 50 foldername` 50 characters between files.
 `ls ?o*` list anything with O as second letter in name.
    
- `grep`: filter
- `pwd` : Print working directory
- `curl`: Download URL
- `pip install bash_kernel`: Installing kernel to write directly on bash
- `unzip`
- `man`: manual
 > If any doubt go to `ls --help`

### Dir-files

- `mkdir`: make directory
- `mv`: move or rename
- `cp`: copy
- `rm`: remove. (`rm -r anything`: delete anything. There's no way to recover whatsoever)
- `rmdir`: remove directory only if empty. 
- `find`: Specify where (`.`if here, `..`if previous dir), `-type`f (file) d (directory), `-name` (`*.pdf`all pdfs in it), `-size` for less than 1Mb (`*1m`)


> `*` all files
