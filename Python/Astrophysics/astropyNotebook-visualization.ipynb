```python

import matplotlib.pyplot as pl
import numpy as np
from astropy.coordinates import SkyCoord
from astropy import units as u
```

```python
Name='M 51'
Fuente = SkyCoord.from_name(Name)  # poner nombre fuete y te busca en SESAME las coordenadas (hay que tener conexi√≥n a internet)
print('Las cooredanadas de', Name,'son ', Fuente)  # nos da coordenadas ecuatoriales
```

```python
# si no ponerlas a mano, indicando unidades *u.units
#crab = SkyCoord(299.87*u.deg, 40.73*u.deg, frame='icrs')
```

```python
# PYTHON Y SDSS (survey de galaxias) NOS PERMITE BAJARNOS DIRECTAMENTE LAS IMAGENES CORTADAS 
from urllib.parse import urlencode
from urllib.request import urlretrieve
from IPython.display import Image
```

```python
im_size = 12*u.arcmin # get a 12 arcmin square
im_pixels = 1024
urlCutOut = 'http://skyservice.pha.jhu.edu/DR12/ImgCutout/getjpeg.aspx'
query_string = urlencode(dict(ra=Fuente.ra.deg,
                              dec=Fuente.dec.deg,
                              width=im_pixels, height=im_pixels,
                              scale=im_size.to(u.arcsec).value/im_pixels))
url = urlCutOut + '?' + query_string
```

```python
urlretrieve(url, Name+'.jpg')
```
