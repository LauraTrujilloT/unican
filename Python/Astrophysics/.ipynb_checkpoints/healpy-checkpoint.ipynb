```python
import healpy as hp
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
map1 = hp.read_map("mapSEVEM.fits")
```

## Un-nested

```python
hp.mollview(map1, min=-2.5e-4, max=2.5e-4)
```

```python
hp.gnomview(map1, rot=[0,0], min=-5e-4, max=5e-4, xsize=500,ysize=500, title="")
```

## Nested

```python
mapNest = hp.reorder(map1, inp="RING", out="NESTED")
hp.mollview(mapNest, min=-5e-4, max=5e-4, nest=True)
```

```python
hp.gnomview(mapNest, rot=[0,0], min=-5e-4, max=5e-4, xsize=500,ysize=500, title="", nest=True)
```

```python
hp.write_map("map-nest3.fits",mapNest, nest=True)
```

```python
np.savetxt("map-nest.ascii", mapNest)
```

#### Filter maps

```python
filtermap1 = hp.smoothing(map1, fwhm= np.deg2rad(1))
filtermap2 = hp.smoothing(map2, fwhm= np.deg2rad(1))
```

```python
plt.figure(figsize=(10,10))
hp.mollview(filtermap2, min=-5e-4, max=5e-4, sub=(1,2,1), title="Filter Commander")
hp.mollview(filtermap1, min=-5e-4, max=5e-4, sub=(1,2,2), title="Filter SEVEM")
```

#### Masks

```python
resolution = hp.nside2resol(1024)
```

```python
phi = np.arange(0, 2 * np.pi, resolution) # radians
theta = np.arange(np.deg2rad(-20) + 0.5 * np.pi, np.deg2rad(20) + 0.5 * np.pi, resolution)
len(theta)
```

```python
len(phi)
```

```python
Nside = 1024
pix = []
for i in phi:
    pixels = hp.ang2pix(Nside, theta=theta, phi=i)
    pix.append(pixels)
```

```python
Nmap1 = hp.read_map("mapSEVEM.fits")
Nmap2 = hp.read_map("COM_CMB_IQU-commander_1024_R2.02_full.fits")
```

```python
map1[pix] = hp.UNSEEN
map2[pix] = hp.UNSEEN
```

```python
plt.figure(figsize=(10,10))
hp.mollview(map2, min=-5e-4, max=5e-4, sub=(1,2,1), title="Mask Commander")
hp.mollview(map1, min=-5e-4, max=5e-4, sub=(1,2,2), title="Mask SEVEM")
```

```python
nside = 1024
cl_map1 = hp.anafast(map1, lmax = 3 * nside)
cn_map1 = hp.anafast(Nmap1, lmax = 3 * nside)
cl_map2 = hp.anafast(map2, lmax = 3 * nside)
cn_map2 = hp.anafast(Nmap2, lmax = 3 * nside)
```

```python
cf_map1 = hp.anafast(filtermap1, lmax = 3 * nside)
cf_map2 = hp.anafast(filtermap2, lmax = 3 * nside)
```

```python
plt.figure(figsize=(16,14))
plt.subplot(2,2,1)
plt.loglog(cl_map1, label="Mask")
plt.loglog(cn_map1, label="Non-mask", color="black")
plt.loglog(cf_map1, label="Filter", color="red")
ax = plt.gca()
#ax.set_xlim([0, 200])
#ax.set_yscale("log")
plt.title("Sevem Map", fontsize=15)
plt.xlabel(r"$l$", fontsize=14)
plt.ylabel(r"$C_{l}$", fontsize=14)
plt.grid()
plt.legend()

plt.subplot(2,2,2)
plt.loglog(cl_map2, label="Mask")
plt.loglog(cn_map1, label="Non-mask", color="black")
plt.loglog(cf_map1, label="Filter", color="red")
plt.title("Commander Map", fontsize=15)
plt.xlabel(r"$l$", fontsize=14)
plt.ylabel(r"$C_{l}$", fontsize=14)
plt.legend()
plt.grid()

plt.show()
```
