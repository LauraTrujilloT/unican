`pyplot` provides a procedural interface to the matplotlib object-oriented plotting library. It is modeled closely after _Matlab_ analogs with similar arguments. Important commands are explained with interactive examples.

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
x = np.linspace(- np.pi, np.pi, 256, endpoint=True) #endpoint stops at last point of x
c, s = np.cos(x), np.sin(x)

plt.figure()
plt.plot(x, c) #Creating plot of cos(x)
plt.plot(x, s)
plt.show()
```

### Ok... Let's customize a bit.

```python
plt.figure(figsize=(8, 6), dpi=80) # Creates a fig of size 8x6 inches, 80 dots per inch
plt.subplot(1, 1, 1) # Creates a fig from grid 1x1
plt.plot(x, c, color="blue", lw=1.0, ls="-") #lw short for linewidth
plt.plot(x, s, color="green", lw=1.0, ls="-") # ls short for linespace
plt.xlim(x.min() * 1.1, x.max() * 1.1) #limits of x : this avoid any overlap with fig and curve
plt.ylim(c.min() * 1.1, c.max() * 1.1)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi]) #manipulating how many ticks I want in the plot
plt.yticks([-1, 0, 1])
plt.title("Changing: Figsize, xlim")
#plt.savefig("fig1.png", dpi=70)
plt.show()
```

### Label axis with LateX

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(1, 1, 1)
plt.plot(x, c, color="blue", lw=1.0, ls="-") 
plt.plot(x, s, color="green", lw=1.0, ls="-")
plt.xlim(x.min() * 1.1, x.max() * 1.1) 
plt.ylim(c.min() * 1.1, c.max() * 1.1)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],
          [r'$-1$', r'$0$', r'$+1$'])
plt.title("Changing: Figsize, xlim")
plt.show()
```

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(1, 1, 1)
plt.plot(x, c, color="blue", lw=2.0, ls="-", label = r"$\cos(x)$") 
plt.plot(x, s, color="red", lw=2.0, ls="-", label = r"$\sin(x)$")


plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
         [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],
          [r'$-1$', r'$0$', r'$+1$'])

ax = plt.gca()  # gca ='get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.title("Using gca")
plt.legend(loc="upper left")
plt.show()
```

### Annotating some points 

```python
t = 2 * np.pi / 3

plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(1, 1, 1)
plt.plot(x, c, color="blue", lw=2.0, ls="-", label = r"$\cos(x)$") 
plt.plot(x, s, color="red", lw=2.0, ls="-", label = r"$\sin(x)$")

# Point
plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=2.5, linestyle="--")
plt.scatter([t, ], [np.cos(t), ], 50, color='blue')

# Annotate:

plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
             xy=(t, np.cos(t)), xycoords='data',
             xytext=(-90, -50), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

#Another point
plt.plot([t, t],[0, np.sin(t)], color='red', linewidth=2.5, linestyle="--")
plt.scatter([t, ],[np.sin(t), ], 50, color='red')

# annotate the point
plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
         [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],
          [r'$-1$', r'$0$', r'$+1$'])

ax = plt.gca()  # gca ='get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.title("Annotating points")
plt.legend(loc="upper left")
plt.show()
```

### Manipulating labels and transparencies

```python
t = 2 * np.pi / 3


plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(1, 1, 1)
plt.plot(x, c, color="blue", lw=2.0, ls="-", label = r"$\cos(x)$") 
plt.plot(x, s, color="red", lw=2.0, ls="-", label = r"$\sin(x)$")

# Point
plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=2.5, linestyle="--")
plt.scatter([t, ], [np.cos(t), ], 50, color='blue')

# Annotate:

plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
             xy=(t, np.cos(t)), xycoords='data',
             xytext=(-90, -50), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

#Another point
plt.plot([t, t],[0, np.sin(t)], color='red', linewidth=2.5, linestyle="--")
plt.scatter([t, ],[np.sin(t), ], 50, color='red')

# annotate the point
plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
         [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 1],
          [r'$-1$', r'$0$', r'$+1$'])

ax = plt.gca()  # gca ='get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65))

plt.title("Annotating points")
plt.legend(loc="upper left", fancybox=True, fontsize=15, shadow=True)
plt.show()
```
